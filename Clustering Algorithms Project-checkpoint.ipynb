# K Means Clustering Project 

This project involves clustering customers based on their age, annual income, and spending score using three different clustering algorithms: K-means, DBSCAN, and Hierarchical clustering.

The dataset used for analysis is named "Customers.csv" and it contains relevant information such as customer ID, gender, age, annual income, and spending score. The aim of this project is to identify distinct groups or clusters of customers based on their characteristics, which can provide valuable insights for targeted marketing strategies.

The focus of the clustering analysis is on the columns 'Age', 'Annual Income (k$)', and 'Spending Score (1-100)'. These columns serve as the basis for clustering the customers.

First, the K-means clustering algorithm is applied to the dataset. K-means aims to partition the data into K clusters, where each data point is assigned to the cluster with the nearest mean. For this project, K-means clustering is performed using the three specified columns.

Next, the Hierarchical clustering algorithm is employed. Hierarchical clustering builds a hierarchy of clusters by successively merging or splitting clusters based on their similarity. The result is a tree-like structure called a dendrogram. In this project, Hierarchical clustering is applied solely to the 'Age' and 'Spending Score (1-100)' columns.

Lastly, the DBSCAN clustering algorithm is utilized. DBSCAN is a density-based clustering algorithm that groups together data points based on their density. Points that are densely connected form clusters, while points with low density are considered noise or outliers. In this project, DBSCAN clustering is conducted on the three columns of interest.



To evaluate the performance of each clustering model, the silhouette score is calculated. The silhouette score measures the cohesion and separation of the clusters, indicating how well-defined and distinct they are from one another. A higher silhouette score indicates better clustering performance.

```python
X = df.iloc[:, [1, 2]].values  # Selects values from columns at index 1 and 2
```

```python
X
```

```python
from sklearn.cluster import KMeans
wcss = []

# Loop through number of clusters from 1 to 10
for i in range(1, 11):
    # Initialize the KMeans algorithm with specified number of clusters, k-means++ initialization, and random state
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    
    # Fit the KMeans model to the data
    kmeans.fit(X)
    
    # Append the within-cluster sum of squares (WCSS) to the list
    wcss.append(kmeans.inertia_)

# Plot the number of clusters against the WCSS
plt.plot(range(1, 11), wcss)

# Set the x-axis label to 'Number of Clusters'
plt.xlabel('Number of Clusters')

# Set the y-axis label to 'WCSS'
plt.ylabel('WCSS')

# Set the title of the plot
plt.title('The Elbow Method')

# Display the plot
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)  # Create a KMeans object with 5 clusters
y_kmeans = kmeans.fit_predict(X)  # Fit the KMeans model to the data and assign cluster labels
print(y_kmeans)  # Print the cluster labels assigned to each data point
```

```python
# Scatter plot of data points in Cluster 1
plt.scatter(df[y_kmeans==0]['Annual Income (k$)'], df[y_kmeans==0]['Spending Score (1-100)'], s=100, c='red', label='Cluster 1')

# Scatter plot of data points in Cluster 2
plt.scatter(df[y_kmeans==1]['Annual Income (k$)'], df[y_kmeans==1]['Spending Score (1-100)'], s=100, c='blue', label='Cluster 2')

# Scatter plot of data points in Cluster 3
plt.scatter(df[y_kmeans==2]['Annual Income (k$)'], df[y_kmeans==2]['Spending Score (1-100)'], s=100, c='green', label='Cluster 3')

# Scatter plot of data points in Cluster 4
plt.scatter(df[y_kmeans==3]['Annual Income (k$)'], df[y_kmeans==3]['Spending Score (1-100)'], s=100, c='cyan', label='Cluster 4')

# Scatter plot of data points in Cluster 5
plt.scatter(df[y_kmeans==4]['Annual Income (k$)'], df[y_kmeans==4]['Spending Score (1-100)'], s=100, c='magenta', label='Cluster 5')

# Scatter plot of cluster centers (centroids)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='yellow', label='Centroids')

# Set the title of the plot
plt.title('Clusters of Customers')

# Set the x-axis label
plt.xlabel('Annual Income (k$)')

# Set the y-axis label
plt.ylabel('Spending Score (1-100)')

# Display the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```

```python
from sklearn.metrics import silhouette_score  # Import silhouette_score function

silhouette_avg = silhouette_score(X, y_kmeans)  # Calculate average silhouette score
print("The average silhouette score is:", silhouette_avg)  # Print the average silhouette score
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values

# Initialize variables for storing the best hyperparameters and silhouette score
best_epsilon = None
best_min_samples = None
best_score = -1

# Iterate over different values of epsilon and min_samples
for epsilon in range(1, 10):
    for min_samples in range(2, 10):
        # Create an instance of DBSCAN with the current hyperparameters
        dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
        
        # Fit the data points to the DBSCAN model and obtain the cluster labels
        labels = dbscan.fit_predict(X)
        
        # Check if the resulting clusters contain at least 2 unique labels
        if len(np.unique(labels)) < 2:
            continue
        
        # Calculate the silhouette score for the current clustering
        score = silhouette_score(X, labels)
        
        # Update the best hyperparameters and silhouette score if the current score is higher
        if score > best_score:
            best_epsilon = epsilon
            best_min_samples = min_samples
            best_score = score

# Print the best hyperparameters and silhouette score
print("Best epsilon:", best_epsilon)
print("Best min_samples:", best_min_samples)
print("Best silhouette score:", best_score)
```

### A brief explanation of this code

- The code loads the dataset from the CSV file and selects the features of interest.
- It then initializes variables to store the best hyperparameters (`best_epsilon`, `best_min_samples`) and the highest silhouette score (`best_score`).
- Nested loops iterate over different values of `epsilon` and `min_samples`.
- Inside the loop, a DBSCAN model is created with the current hyperparameters and fitted to the data points.
- The resulting cluster labels are used to calculate the silhouette score.
- If the number of unique labels in the cluster is less than 2, the iteration continues to the next iteration.
- If the current silhouette score is higher than the previous best score, the hyperparameters and score are updated.
- Finally, the best hyperparameters and silhouette score are printed.
